Quick outlier detection summary

```python
for col in numeric_cols:
    if col != 'index': # Exclude 'index' column
        Q1, Q3 = df[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
        outliers = ((df[col] < lower) | (df[col] > upper)).sum()
        print(f"{col}: {outliers} outliers detected ({100*outliers/len(df):.2f}%)")
```

```python
df1['stops'].value_counts()
```

```python
# 1. Find the smallest class size
min_count = df1['stops'].value_counts().min()

# 2. Downsample each airline group
df = df1.groupby('stops').sample(n=min_count, random_state=42)
```

# 2. Exploratory data **analysis**

Dataset overview visualization

*Distribution of Airlines*

```python
plt.figure(figsize=(10,5))
airline_counts = df['airline'].value_counts()
plt.bar(airline_counts.index, airline_counts.values, color='skyblue', edgecolor='black')
plt.title('Number of Flights per Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Distribution of Flight Classes (Economy vs Business)*

```python
plt.figure(figsize=(6,4))
class_counts = df['class'].value_counts()
plt.bar(class_counts.index, class_counts.values, color=['orange','lightgreen'], edgecolor='black')
plt.title('Distribution of Flight Classes')
plt.xlabel('Flight Class')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Number of Stops*

```python
plt.figure(figsize=(6,4))
stops_counts = df['stops'].value_counts().sort_index()
plt.bar(stops_counts.index.astype(str), stops_counts.values, color='lightcoral', edgecolor='black')
plt.title('Distribution of Number of Stops')
plt.xlabel('Stops')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Ticket Price Distribution*

```python
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=40, color='steelblue', edgecolor='black', alpha=0.8)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Days Left Before Departure (Booking Window)*

*Average Ticket Price per Airline*

```python
plt.figure(figsize=(10,5))
avg_price_airline = df.groupby('airline')['price'].mean().sort_values(ascending=False)
plt.bar(avg_price_airline.index, avg_price_airline.values, color='mediumpurple', edgecolor='black')
plt.title('Average Ticket Price per Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Correlation (numeric columns only)*

```python
df_numeric_cols = df.select_dtypes(include=np.number).drop(columns=['index'])
corr_matrix = df_numeric_cols.corr()
print("\nCorrelation Matrix:\n", corr_matrix)

plt.figure(figsize=(6,4))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Heatmap (Numerical Features)')
plt.tight_layout()
plt.show()
```

*	RQ1: How does the number of days left before departure affect flight ticket
prices?
*	Hypothesis: Flights booked closer to the departure date are significantly more expensive than those booked well in advance.


*Scatter Plot: Days_Left vs Price*

```python
plt.figure(figsize=(8,5))
plt.scatter(df['days_left'], df['price'], alpha=0.4, color='royalblue', edgecolor='black', linewidth=0.5)
plt.title('Scatter Plot: Days Left vs Ticket Price')
plt.xlabel('Days Left Before Departure')
plt.ylabel('Ticket Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Trend Line (Average Price vs Days_Left Binned)*

```python
df1 = df.copy()
df1['Days_Bin'] = pd.cut(df1['days_left'], bins=np.arange(0, df1['days_left'].max()+5, 5))
avg_price_by_bin = df1.groupby('Days_Bin')['price'].mean()

plt.figure(figsize=(10,5))
plt.plot(avg_price_by_bin.index.astype(str), avg_price_by_bin.values, marker='o', color='darkorange', linewidth=2)
plt.title('Average Ticket Price by Booking Window (5-Day Bins)')
plt.xlabel('Days Left (Binned)')
plt.ylabel('Average Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Correlation between Days_Left and Price*

```python
corr = df['days_left'].corr(df['price'])
print(f"\nCorrelation between Days_Left and Price: {corr:.3f}")
if corr < 0:
    print("ðŸ”¹ Negative correlation â€” as departure nears (fewer days left), prices increase.")
elif corr > 0:
    print("ðŸ”¹ Positive correlation â€” as departure nears (fewer days left), prices decrease.")
else:
    print("ðŸ”¹ No clear correlation detected.")
```

*Box Plot: Price Distribution Across Booking Windows*

```python
max_day = df1['days_left'].max()
bins = [0, 5, 10, 20, 30, 60, 90, 120, max(121, max_day)]
labels = ['0-5','6-10','11-20','21-30','31-60','61-90','91-120','120+']

df1['Booking_Window'] = pd.cut(df1['days_left'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(10,5))
df1.boxplot(column='price', by='Booking_Window', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='navy'),
           medianprops=dict(color='red', linewidth=2))
plt.title('Box Plot: Ticket Price Distribution by Booking Window')
plt.suptitle('')
plt.xlabel('Days Left Category')
plt.ylabel('Ticket Price')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*	RQ2: Does the effect of days left on ticket prices vary across different airlines?
* Hypothesis: Airlines differ in their pricing strategies, with some showing sharper increases in ticket prices as departure approaches compared to others.


*Overview: Average Ticket Price by Airline*

```python
plt.figure(figsize=(10,5))
avg_price_airline = df.groupby('airline')['price'].mean().sort_values(ascending=False)
plt.bar(avg_price_airline.index, avg_price_airline.values, color='steelblue', edgecolor='black')
plt.title('Average Ticket Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Scatter Plot: Days_Left vs Price per Airline*

```python
unique_airlines = df['airline'].unique()
n_airlines = len(unique_airlines)
cols = 3
rows = int(np.ceil(n_airlines / cols))

plt.figure(figsize=(16, 4*rows))
for i, airline in enumerate(unique_airlines, 1):
    plt.subplot(rows, cols, i)
    subset = df[df['airline'] == airline]
    plt.scatter(subset['days_left'], subset['price'], alpha=0.5, s=20, color='teal', edgecolor='black', linewidth=0.3)
    plt.title(f'{airline}')
    plt.xlabel('Days Left')
    plt.ylabel('Price')
    plt.grid(True, linestyle='--', alpha=0.5)
plt.suptitle('Scatter Plot: Days Left vs Price by Airline', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

*Trend Lines (Average Price per Days_Left Binned per Airline)*

```python
# Bin Days_Left into 5-day windows
df1['Days_Bin'] = pd.cut(df1['days_left'], bins=np.arange(0, df1['days_left'].max()+5, 5))

plt.figure(figsize=(12,6))
for airline in df1['airline'].unique():
    subset = df1[df1['airline'] == airline]
    avg_price_by_bin = subset.groupby('Days_Bin')['price'].mean()
    plt.plot(avg_price_by_bin.index.astype(str), avg_price_by_bin.values, marker='o', linewidth=2, label=airline)

plt.title('Average Ticket Price by Booking Window for Each Airline')
plt.xlabel('Days Left (5-Day Bins)')
plt.ylabel('Average Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Airline', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Combined Trend: Days_Left vs Price for Airlines (Economy Only for Clarity)*

```python
economy_df = df1[df1['class'].str.lower() == 'economy']

# Bin days left
economy_df['Days_Bin'] = pd.cut(economy_df['days_left'], bins=np.arange(0, economy_df['days_left'].max()+5, 5))

plt.figure(figsize=(12,6))
for airline in economy_df['airline'].unique():
    subset = economy_df[economy_df['airline'] == airline]
    avg_price_bin = subset.groupby('Days_Bin')['price'].mean()
    plt.plot(avg_price_bin.index.astype(str), avg_price_bin.values, marker='o', linewidth=2, label=airline)

plt.title('Average Price by Booking Window (Days_Left) across Airlines (Economy Class)')
plt.xlabel('Days Left (5-Day Bins)')
plt.ylabel('Average Ticket Price')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Airline', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

*Multi-Factor Correlation Summary (Airline Ã— Class Ã— Stops)*

```python
corr_summary = (
    df.groupby(['airline', 'class', 'stops'])
      .apply(lambda x: x['days_left'].corr(x['price']))
      .reset_index(name='Correlation')
)

plt.figure(figsize=(12,6))
airlines = corr_summary['airline'].unique()
for i, airline in enumerate(airlines):
    subset = corr_summary[corr_summary['airline'] == airline]
    plt.plot(subset['stops'], subset['Correlation'], marker='o', label=airline)

plt.title('Correlation (Days_Left vs Price) across Airlines and Stops (All Classes Combined)')
plt.xlabel('Number of Stops')
plt.ylabel('Correlation Coefficient')
plt.legend(title='Airline', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("Correlation Summary by Airline, Class, and Stops")
display(corr_summary.round(3))
```
